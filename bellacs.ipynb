# ESERCIZI FATTI USANDO FILE SAMPLING

## 5. For each station (itcs_stopName), determine the average number of passengers.

```python
import os # importato solo perche sto usando il file sampling
df_stops = pd.DataFrame()
dtype_mapping = {'itcs_numberOfPassengers': float, 'itcs_stopName': str}
for name in dati['name']:
    file_path= f'ZTBus_samples/{name}.csv'
    if os.path.exists(file_path):
        df = pd.read_csv(file_path, na_values='-',usecols=['itcs_numberOfPassengers','itcs_stopName'],dtype=dtype_mapping).dropna(subset=['itcs_stopName'])
        df_stops=pd.concat([df_stops,df])
```
